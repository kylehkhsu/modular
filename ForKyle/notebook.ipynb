```python
import numpy as np
import matplotlib.pyplot as plt
# import utilities as utils
import importlib
import metrics
importlib.reload(metrics)
from metrics import compute_linear_metrics
from pprint import pprint

dropouts = [0, 1]
data_distributions_3 = [np.load(f'./data_distribution_dropout_{dropout}.npy') for dropout
                        in dropouts]
activity_data_3 = [np.load(f'./responses_dropout_{dropout}.npy', allow_pickle=True) for dropout in dropouts]

```

```python
for i, dropout in enumerate(dropouts):
    for j, response in enumerate(activity_data_3[i]):
        print(f'Dropout: {dropout}, Model: {j+1}')
        activations = np.array(response['activations']).reshape(-1, 25)
        inputs_info = response['input']

        sources = np.zeros((81, 2))
        latents = np.zeros((81, 25))

        for idx, info in enumerate(inputs_info):
            sources[idx, 0] = info['what_index']
            sources[idx, 1] = info['where_index']
            if data_distributions_3[i][info['what_index']][info['where_index']] == 1:
                latents[idx] = activations[idx]
            else:
                latents[idx] = np.nan
        sources = sources[~np.isnan(latents).any(axis=1)]
        latents = latents[~np.isnan(latents).any(axis=1)]
        metrics = compute_linear_metrics(sources, latents, 'continuous')
        pprint(metrics)
```
