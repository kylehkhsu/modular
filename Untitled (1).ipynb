```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s1 = [2,0,0,2,2,1,1,1]

s2 = [0,1,2,1,2,2,0,1]


s_pairs = [(x,y) for x,y in zip(s1,s2)]

def set_ax(ax):
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    
```

```python
fig, ax = plt.subplots(1,1, figsize=(2,2), dpi = 300)
set_ax(ax)
ms = 3
ax.plot([0,1], [1,0], c = 'C1', ms=ms)
ax.fill_between(x=[0,1], y1 = [1,0], alpha = 0.3, hatch = '/', color = 'C1')
ax.scatter(s1, s2, c= 'gray', zorder = 10)
ax.set_xlabel('Shape')
ax.set_ylabel('Color')
fig.savefig("corner.png", transparent = True, pad_inches = 0.2)
```

```python
fig, ax = plt.subplots(1,1, figsize=(2,2), dpi = 300)
set_ax(ax)
c_shape= 'tomato'
c_color = 'skyblue'
ax.set_ylim(-0.1,4)
ax.plot(np.arange(len(s1)),np.array(s1), '-o', label = 'Shape', c=c_shape, ms = ms)
ax.plot(np.arange(len(s2)),np.array(s2), '-o', label = 'Color', c= c_color, ms = ms)
ax.set_ylabel('Acitivity')
ax.set_xlabel('Stimulus [i]')
plt.legend(frameon=False,  bbox_to_anchor = [0.8,1])
fig.savefig("feature",  transparent = True, pad_inches = 0.2)
```

```python
u,v = (1., 1)
b1, b2 = (0,0)
modular_x = [u*x + u*b1 for x in s1]
modular_y = [v*y + v*b2  for y in s2]
mixed = [u*x + u*b1 + y*v + y*b2 for x,y in s_pairs]
fig, (ax1, ax2) = plt.subplots(1,2,sharey=True, figsize=(3,1.8), dpi = 300)
#plt.subplots_adjust(wspace=0.3)
set_ax(ax1)
set_ax(ax2)
l1,=ax1.plot(modular_x, '-o', c =c_shape, label = 'Shape', ms = ms)
ax1.fill_between(np.arange(len(modular_x)), modular_x, color = c_shape, alpha = 0.3, hatch= '//', label = 'Shape')
ax1.set_ylabel('Activity ')
ax1.set_xlabel('Stimulus [i]')
ax1.set_ylim(-0.1,3)
#ax1.legend(handles= [l1], frameon=False, loc = "upper right")

#bias = np.mean([x*y for (x,y) in s_pairs])
#mixed_solution = modular_solution - bias

ax2.set_xlabel('Stimulus [i]')
c_mixed = 'yellowgreen'
l2,=ax2.plot(modular_y, '-o', c= c_color, label = 'Color', ms = ms)
ax2.fill_between(np.arange(len(modular_y)), modular_y, color = c_color, alpha = 0.3, hatch= '//')
ax2.set_ylim(-0.1,4)
#ax2.legend(handles= [l2], frameon=False,loc = "upper right")
#plt.legend(handles = [l1,l2], frameon = False)

#ax2.fill_between(np.arange(len(modular_solution)), mixed_solution, color= c_mixed, alpha = 0.3, hatch = '//')
#ax2.fill_between(np.arange(len(modular_solution)), modular_solution, mixed_solution, color= 'yellow', alpha = 0.3)

fig.savefig("modular.png", transparent = True, pad_inches = 0.2)
```
